```python
coefs = np.polynomial.polynomial.polyfit(t, y, 2)
ffit = np.poly1d(coefs)
t_new = np.linspace(t[0], t[-1], num=len(t)*10)
fig1 = plt.figure()                                                                                           
at1 = fig1.add_subplot(111)                                                                                   
at1.scatter(t, y, facecolors='None')                                                                     
at1.plot(t_new, ffit(t_new))                                                                     
plt.show()
```
