```
%matplotlib inline
import pylab as pl
import numpy as np
#Función por partes f1
def f1_parts(x):
  if x<0:
    y=x+2
  if 0<=x<=1:
    y=2
  if x>1:
    y=-x+3
  return y

#Proceso Grafica F1
x0=float(input("Ingrese el primer valor a tomar en x: "))
xp=float(input("Ingrese el paso: "))
xf=float(input("Ingrese el valor final de x: "))
dominio=np.arange(x0,xf+1,xp)
rango=np.array([f1_parts(xi) for xi in dominio])
pl.plot(dominio,rango)
pl.show()
```

```
import pylab as pl
#Funcion parte entera
def floor(x):
  if x>=0:
    y=x//1
  else:
    y=(x//1)
  return y

def ceil(x):
  if x>=0:
    y=abs(x//-1)
  else:
    y=-(x//-1)
  return y

x0=float(input("Ingrese el primer valor a tomar en x: "))
xp=float(input("Ingrese el paso: "))
xf=float(input("Ingrese el valor final de x: "))
f=eval(input("ceil(0) or floor(1): "))
#Grafico
if f==0:
  dom=np.arange(x0,xf,xp)
  rango=np.array([ceil(i) for i in dom])
  pl.title(r"$\left \lceil{x}\right \rceil$")
  pl.scatter(dom,rango)

elif f==1:
  dom=np.arange(x0,xf,xp)
  rango=np.array([floor(i) for i in dom])
  pl.title(r"$\left \lfloor{x}\right \rfloor$")
  pl.scatter(dom,rango)
else:
  print("Ingrese una opción correcta")
```

```
%matplotlib inline
#QUIZ 3
import numpy as np
import pylab as pl

def f1(x,a,b):
  if 0<x<=a:
    y=5
  elif a<x<=a+b:
    y=0
  elif a+b<x<=2*a+b:
    y=5
  elif 2*a+b<x<=2*a+2*b:
    y=0
  elif 2*a+2*b<x<=3*a+2*b:
    y=5
  elif 3*a+2*b<x<=3*a+3*b:
    y=0
  else:
    y=0
  return y

dominio=np.arange(0,15,0.1)
rango=np.array([f1(xi,3,2) + np.sin(10*xi) for xi in dominio])

pl.plot(dominio,rango)
pl.show()


```

```
#Funciones Gaussianas
import numpy as np
import pylab as pl
def gaussF(x,μ,σ):
  gF=np.exp((-1/2)*((x-μ)/σ)**2)/(σ*np.sqrt(2*np.pi))*np.sin(5*x)
  return gF

x=np.arange(-5,5,0.01)
a=np.arange(0,10)
for i in a:
  y=np.array([gaussF(xi,i,i/1.2) for xi in x])
  pl.plot(x,y,label=r"gF")
  pl.legend()
  
```
