```python
import pandas as pd
REORDER=True
df=pd.read_csv('https://raw.githubusercontent.com/ComputationalMethods/Evaluacion_2020-2/master/Calificaciones.csv')
df['T']=(((df.T1+df.T2+df.T3+df.T4+df.T5+df.T6)/6+df.TA)/2).round(1)
df['F']=(df.E1+df.E2+df.E3+df.TF+df['T'])/5
if REORDER:
    df=df.iloc[[8]].append(df.iloc[:8] ).append( df.iloc[9:] ).reset_index(drop=True)
df[['Cédula','E1','E2','E3','TF','T','F']].fillna(0)
```
