```python
for i in range(1,4):
    df[f'E{i}']=df[f'Parcial_{i}'].str['value']
```

```python
df['TF']=df['Trabajo_Final'].str['value']
```

```python
for i in range(3):
    df[f'T{i+1}']=df['Tareas'].str[i].str['value']
```

```python
df['TA']=df['Tareas'].apply(lambda l: [d for d in l if d.get('Taller')]).str[0].str['value'].fillna(0.)
```

```python
df['Cédula']=df['student_id']
```

```python
dff=df[['Cédula','E1','E2','E3','TF','T1','T2','T3','TA']]
```

```python
dff=tm[['Cédula']].merge(dff,on='Cédula',how='left').fillna(0)
dff.to_csv('calificaciones.csv',index=False)
```

```python
import pandas as pd
REORDER=True
df=pd.read_csv('https://github.com/ComputationalMethods/Evaluacion_2021-2/raw/main/calificaciones.csv')
df['T']=((  (df.T1+df.T2+df.T3)/3*0.25+df.TA*0.75   )).round(1)
df['F']=(df.E1+df.E2+df.E3+df.TF+df['T'])/5
if REORDER:
    df=df.iloc[[8]].append(df.iloc[:8] ).append( df.iloc[9:] ).reset_index(drop=True)
df[['Cédula','E1','E2','E3','TF','T','F']].fillna(0)
```

```python
df.to_csv('calificaciones.csv',index=False)
```
