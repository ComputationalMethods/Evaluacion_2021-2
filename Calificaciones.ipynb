```python
df=pd.read_json('https://raw.githubusercontent.com/ComputationalMethods/Evaluacion_2021-2/main/calificaciones.json',dtype={'student_id':str})
```

```python
df=df[df['student_id']!='0000000000'].reset_index(drop=True)
```

download from MARES

```python
tm=pd.read_csv('listado.csv',dtype={'Cédula':str})
```

```python
for i in range(1,4):
    df[f'E{i}']=df[f'Parcial_{i}'].str['value']
```

```python
df['TF']=df['Trabajo_Final'].str['value']
```

```python
for i in range(3):
    df[f'T{i+1}']=df['Tareas'].str[i].str['value']
```

```python
df['TA']=df['Tareas'].apply(lambda l: [d for d in l if d.get('Taller')]).str[0].str['value'].fillna(0.)
```

```python
df['Cédula']=df['student_id']
```

```python
dff=df[['Cédula','E1','E2','E3','TF','T1','T2','T3','TA']]
```

```python
dff=tm[['Cédula']].merge(dff,on='Cédula',how='left').fillna(0)
dff.to_csv('calificaciones.csv',index=False)
```
